### 3.1 Perfilamiento fuentes de datos #1

La primera fuente de datos contiene únicamente fotos de personas que no están usando tapaboca. La estructura de la data set consiste en una lista de carpetas por sujeto, donde dentro hay 1 o más fotos de la cara de la persona.

```python
# Función que permite leer la data set descrita anteriormente dadas las columnas para guardar la información relevante
# y el directorio donde se encuentra
def readDataSetWithSubjects(directory, widthColumn, heightColumn, dirColumn, sizeMemoryColumn, subjectColumn, formatColumn):
    
    # Iterate over the directories in the data set
    files = os.listdir(directory)
    for f in files:
        
        # Directory from each person
        dirPerson = directory + f

        #Iterate over the images of each person
        for i in os.listdir(dirPerson):
            #Dir of an image of the person
            dirImg = dirPerson + '/' + i
        
            #Get width, height, size
            img = Image.open(dirImg)
            width, height = img.size
            size = os.stat(dirImg).st_size
            
            # Get the format
            formatOfImg = dirImg.split(".")[-1]
        
            #Append the data to the arrays
            widthColumn.append(width)
            heightColumn.append(height)
            sizeMemoryColumn.append(size)
            dirColumn.append(dirImg)
            subjectColumn.append(f)
            formatColumn.append(formatOfImg)
            
 
```

Cabe destacar que el el objetivo de esta data set originalmente se enfocaba en face recognition, por lo cual dentro de cada carpeta hay subcarpetas que representan a una persona, y dentro de dicha se encuentran las fotos del sujeto

```python
# Función que permite leer la data set descrita anteriormente dadas las columnas para guardar la información relevante
# y el directorio donde se encuentra
def readDataSetWithSubjectsDataSet2(directory, widthColumn, heightColumn, dirColumn, sizeMemoryColumn, subjectColumn, formatColumn, labels):
    
    #Iterate over both directories
    dirs = os.listdir(directory)
    
    # First directory is the one with out face masks
    maskDirectory = False
    for subdirectory in dirs:
        
        #Cur dir
        curDir = directory + subdirectory
        
        # Iterate over the directories in the data set
        files = os.listdir(curDir)
        for f in files:

            # Directory from each person
            dirPerson = curDir + '/' + f

            #Iterate over the images of each person
            for i in os.listdir(dirPerson):
                #Dir of an image of the person
                dirImg = dirPerson + '/' + i

                #Get width, height, size
                img = Image.open(dirImg)
                width, height = img.size
                size = os.stat(dirImg).st_size

                # Get the format
                formatOfImg = dirImg.split(".")[-1]
                
                # Get label
                label = "with_face_mask"
                if not maskDirectory:
                    label = "not_face_mask"

                #Append the data to the arrays
                widthColumn.append(width)
                heightColumn.append(height)
                sizeMemoryColumn.append(size)
                dirColumn.append(dirImg)
                subjectColumn.append(f)
                formatColumn.append(formatOfImg)
                labels.append(label)
        
        # Entering second directory (face masks)
        maskDirectory = True
```

La simulación de imágenes con uso de tapabocas incorrecto se basará en la imagen publicada por la revista Vale indicando los errores más comunes al momento de usar tapabocas:
<img src="http://www.vale.com/brasil/pt/aboutvale/news/publishingimages/news/2020/maio/mascara/header-mas-desk.png" ><br>

Para procesar las imágenes y simular el uso de tapabocas, se realiza un archivo .py que usa el paquete 'face_recognition' y obteniendo los puntos clave de la cara (naríz y mentón) se hace el siguiente procedimiento por cada foto:

1) Se indica el path de la imagen con la cara, path de la máscara a sobreponer, el path en donde debe ser guardada la imagen a ser construída y si la máscara debe ser puesta correctamente o incorrectamente. 

2) Detectar si en la foto se encuentra la cara de una persona usando el paquete face_recognition, si no se encuentra ninguna cara se detiene la ejecución para esa foto.

3) Si hay una cara de una persona, por medio del paquete face_recognition se obtienen los pixeles donde se encuentra la punta de la nariz, los extremos del mentón y la punta del mentón. Con estos puntos se calcula cómo debe rotarse la máscara al momento de sobreponerla.

4) Con los puntos anteriores se utiliza PIL para poner la máscara indicada (en formato .png) sobre la cara de la persona. Si se indica que la máscara debe ser puesta 'incorrectamente' se sobrepone por debajo de la nariz, de lo contrario desde arriba de la punta de la nariz hasta la punta del mentón

* El código utilizado para poner las máscaras fue principalmente obtenido del siguiente repositorio: https://github.com/prajnasb/observations, donde se modificó y añadió lo siguiente:
 - Iterar sobre los datos mostrados en perfilamiento
 - Funcionalidad de poner la máscara 'usada incorrectamente'
 - Funcionalidad para poner cualquier máscara dada por parámetro


* La ejecución de este proceso se hace en un .py ya que no fue posible instalar el packete 'face_recognition' en el ambiente de python de anaconda

### 2.3 Limpieza manual sobre los resultados de la simulación
- Se miran todas las imágenes simuladas manualmente y se eliminan las que por variadas razones como expresiones de las personas, objetos cruzados, tamaño de la cara en pixeles al sobreponerse el tapabocas no daban resultado esperado.

Los estandáres para decidir cuando eliminar fueron:
- Tiras demasiado alejadas de las orejas

### 2.4 Separar train y test
Se construye un dataframe con directorios y clase, luego se crea un .csv con esos datos

```python
directories = []
classCol = []

#Iterate over processed data
dir = 'ProcessedData/'
files = os.listdir(dir)
for i in files:
    clase = i
    if not clase == 'data.csv' and not clase == 'splitedProcessedData': 
        for j in os.listdir(dir + i):
            directories.append(dir + i + '/' + j)
            classCol.append(clase)

# Iterate over the cropped images from data set #3
dir = 'OriginalData/kaggle_andrewmvd/imagesCropped/'
files = os.listdir(dir)
for i in files:
    clase = i
    for j in os.listdir(dir + i):
        directories.append(dir + i + '/' + j)
        classCol.append(clase)   
            
#Build dataframe
dataTable ={ 
    'Directory':  directories,
    'Class': classCol,    
}
table = pd.DataFrame(dataTable, columns = ['Directory', 'Class'])
table        
```
